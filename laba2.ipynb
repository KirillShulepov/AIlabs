**Задание 3**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('dataset.csv')

sns.countplot(x='Target', data=data)
plt.title('Распределение целевой переменной')
plt.show()

print(f"Количество уникальных значений целевой переменной: {data['Target'].nunique()}")
```

Целевая переменная дискретна и принимает три значения: "Dropout", "Graduate", "Enrolled".

**Задание 8**

```python
sns.boxplot(x='Target', y='Age at enrollment', data=data)
plt.title('Зависимость возраста при поступлении от целевой переменной')
plt.show()

sns.boxplot(x='Target', y='Father\'s qualification', data=data)
plt.title('Зависимость уровня образования отца от целевой переменной')
plt.show()

sns.boxplot(x='Target', y='Mother\'s qualification', data=data)
plt.title('Зависимость уровня образования матери от целевой переменной')
plt.show()

sns.boxplot(x='Target', y='Curricular units 2nd sem (grade)', data=data)
plt.title('Зависимость среднего балла за второй семестр от целевой переменной')
plt.show()

sns.boxplot(x='Target', y='Curricular units 1st sem (grade)', data=data)
plt.title('Зависимость среднего балла за первый семестр от целевой переменной')
plt.show()
```

График показывает, что студенты, которые продолжают обучение (Enrolled), как правило, моложе, чем те, кто был отчислен (Dropout) или успешно завершил обучение (Graduate).

График показывает, что студенты, которые успешно завершили обучение (Graduate), имеют более высокие средние баллы за первый семестр, чем те, кто был отчислен (Dropout).

График показывает, что студенты, которые успешно завершили обучение (Graduate), имеют более высокие средние баллы за второй семестр, чем те, кто был отчислен (Dropout).

График показывает, что студенты, которые успешно завершили обучение (Graduate), как правило, имеют матерей с более высоким уровнем образования.

График показывает, что студенты, которые успешно завершили обучение (Graduate), как правило, имеют отцов с более высоким уровнем образования.
