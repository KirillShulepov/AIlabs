Целевая переменная — Target, которая может принимать значения "Dropout", "Graduate" или "Enrolled". Это задача классификации, где мы предсказываем один из трех классов.

**Задание 3**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Загрузка данных
data = pd.read_csv('dataset.csv')

# Гистограмма распределения целевой переменной
sns.countplot(x='Target', data=data)
plt.title('Распределение целевой переменной')
plt.show()

# Количество уникальных значений целевой переменной
print(f"Количество уникальных значений целевой переменной: {data['Target'].nunique()}")
```

Целевая переменная дискретна и принимает три значения: "Dropout", "Graduate", "Enrolled".

Распределение может быть несбалансированным, что может повлиять на точность модели.

**Задание 8**

```python
# Графики зависимости
sns.boxplot(x='Target', y='Age at enrollment', data=data)
plt.title('Зависимость возраста при поступлении от целевой переменной')
plt.show()

sns.boxplot(x='Target', y='Unemployment rate', data=data)
plt.title('Зависимость уровня безработицы от целевой переменной')
plt.show()

sns.boxplot(x='Target', y='Inflation rate', data=data)
plt.title('Зависимость уровня инфляции от целевой переменной')
plt.show()

sns.boxplot(x='Target', y='GDP', data=data)
plt.title('Зависимость ВВП от целевой переменной')
plt.show()

sns.boxplot(x='Target', y='Curricular units 1st sem (grade)', data=data)
plt.title('Зависимость оценки за первый семестр от целевой переменной')
plt.show()
```

Возраст при поступлении может быть связан с целевой переменной, так как студенты разных возрастов могут иметь разные результаты (Dropout, Graduate, Enrolled).

Уровень безработицы и инфляции могут влиять на решение студентов о продолжении обучения.

ВВП может быть связан с экономическими условиями, которые влияют на студентов.

Оценка за первый семестр может быть хорошим предиктором для прогнозирования целевой переменной.

**Задание 9**

```python
# Зависимость двух переменных от целевой переменной
sns.scatterplot(x='Age at enrollment', y='Curricular units 1st sem (grade)', hue='Target', data=data)
plt.title('Зависимость возраста при поступлении и оценки за первый семестр от целевой переменной')
plt.show()
```

На графике видно, что студенты разных возрастов и с разными оценками могут иметь разные результаты по целевой переменной.

Это может быть полезным для построения модели прогнозирования.

```python
# Heatmap для категориальных признаков
sns.heatmap(pd.crosstab(data['Target'], data['Marital status']), annot=True, fmt='d', cmap='YlGnBu')
plt.title('Heatmap зависимости целевой переменной от семейного положения')
plt.show()
```

Heatmap показывает, как семейное положение студентов связано с их результатами (Dropout, Graduate, Enrolled).

Это может помочь в выявлении закономерностей и построении модели.
